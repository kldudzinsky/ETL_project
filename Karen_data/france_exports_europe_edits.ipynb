```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import .csv file
cheese1_file = "france_exports_europe.csv"
cheese1_df = pd.read_csv(cheese1_file)
cheese1_df.head()
```

```python
#drop column called 'DAIRYPROD' as it is not needed here
cheese1_df.drop(columns=['DAIRYPROD'], inplace=True)
cheese1_df.head()
```

```python
#Drop column called 'GEO' as it is not needed here
cheese1_df.drop(columns=['GEO'], inplace=True)
cheese1_df.head()
```

```python
#Drop column called 'Flag and Footnotes' as it is not needed here
cheese1_df.drop(columns=['Flag and Footnotes'], inplace=True)
cheese1_df.head()
```

```python
#rename columns
cheese1_df.columns = ['dairyprod_label','time','milk_item','value','milk_type']
cheese1_df.head()
```

```python
#export file as .csv for use in PostGres
cheese1_df.to_csv('france_exports_europe_fin.csv',index = False)
```
