```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Export chocolate csv into Dataframes
chocolate_file = "Resources/chocolate.csv"
chocolate_df = pd.read_csv(chocolate_file)
chocolate_df.head()
```

```python
#Clean dataset for only companies located in France
chocolate_df = chocolate_df.loc[chocolate_df['company_location']=='France']
chocolate_df.reset_index(inplace=True)
chocolate_df.drop(['index'], axis='columns', inplace=True)
chocolate_df['chocolate_id'] = chocolate_df.index + 1
chocolate_df.head()
```

```python
#Create and Transform Chocolate Company Dataset
chocolate_companies = chocolate_df.company.unique()
chocolate_companies_df = pd.DataFrame({'company': chocolate_companies})
chocolate_companies_df["company_id"] = chocolate_companies_df.index + 1
chocolate_companies_df = chocolate_companies_df[['company_id','company']]
chocolate_companies_df.head()
```

```python
#Create and Transform Chocolate Rating Dataset
chocolate_rating = chocolate_df[['chocolate_id','company','review_date','rating']]
chocolate_rating = chocolate_rating[['chocolate_id','company','review_date','rating']]
chocolate_rating.head()
```

```python
#Create and Transform Chocolate Ingredients Dataset
chocolate_ingredients = chocolate_df[['chocolate_id','company','country_of_bean_origin','cocoa_percent','cocoa_butter','vanilla','lecithin','salt','sugar','sweetener_without_sugar']]
b = {'have_cocoa_butter': True, 'have_not_cocoa_butter': False, 'have_vanila': True, 'have_not_vanila': False,'have_lecithin': True, 'have_not_lecithin': False,'have_salt': True, 'have_not_salt': False,'have_sugar': True, 'have_not_sugar': False,'have_sweetener_without_sugar': True, 'have_not_sweetener_without_sugar': False,}
chocolate_ingredients_df = chocolate_ingredients.replace({'cocoa_butter': b, 'vanilla': b, 'lecithin':b, 'salt':b,'sugar':b,'sweetener_without_sugar':b,})
chocolate_ingredients_df.head()
```

```python
#Setting the index for ID
chocolate_companies_df.set_index("company_id", inplace=True)
chocolate_rating.set_index("chocolate_id", inplace=True)
chocolate_ingredients_df.set_index("chocolate_id", inplace=True)
```

```python
#Connect to Database
connection_string = "postgres:5411sugar@localhost:5432/chocolate"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
chocolate_companies_df.to_sql(name='chocolate_company', con=engine, if_exists='append', index=True)
```

```python
chocolate_rating.to_sql(name='chocolate_rating', con=engine, if_exists='append', index=True)
```

```python
chocolate_ingredients_df.to_sql(name='chocolate_ingredients', con=engine, if_exists='append', index=True)
```
